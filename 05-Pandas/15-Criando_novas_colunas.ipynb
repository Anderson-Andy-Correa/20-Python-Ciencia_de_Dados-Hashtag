```python
import pandas as pd

BASE = r"./dados/cereal_sem_negativos.csv"

df = pd.read_csv(BASE)

df.head()
```

```python
if not "total_carb" in df.columns:
    # df["total_carb"] = df["carbo"] + df["sugars"]
    # Insert total carb before the existing column carb
    df.insert(df.columns.get_loc("carbo"), "total_carb", df["carbo"] + df["sugars"])
df.head()
```

```python
if not "weight_grams" in df.columns:
    df.insert(df.columns.get_loc("weight") + 1, "weight_grams", df["weight"] * 28.3495)

df.head()
```

```python
with pd.option_context("float_format", "{:.2f}".format):
    display(df.describe())
```
