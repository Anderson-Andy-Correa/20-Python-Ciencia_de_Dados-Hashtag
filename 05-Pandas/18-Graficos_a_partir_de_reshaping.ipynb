# Pandas b√°sico

```python
import pandas as pd

BASE = r"./dados/cereal_sem_negativos.csv"

df = pd.read_csv(BASE)

df.head()
```

```python
df.info()
```

```python
df.groupby(["mfr", "shelf"]).count()
```

```python
df.groupby(["mfr", "shelf"]).size().to_frame()
```

```python
df.groupby(["mfr", "shelf"]).size().to_frame().plot(kind="bar")
```

```python
df.groupby(["mfr", "shelf"]).size().unstack("shelf").fillna("-")
```

```python
df.groupby(["mfr", "shelf"]).size().unstack("shelf") \
    .plot(
        kind="bar",
        rot=True,
        title="Quantidade de cereais por pratileira",
        stacked=True, 
        figsize=(10, 5),
        xlabel="Fabricante",
        ylabel="Quantidade",
        )
```

```python
df.groupby(["mfr", "shelf"]).size().unstack().stack()
```

```python
df.pivot_table(
    index="mfr",
    columns="shelf",
    aggfunc="size"
)
```

```python
df.pivot_table(
    index="mfr",
    columns="shelf",
    aggfunc="size"
).plot(kind="bar",
    rot=True,
    title="Quantidade de cereais por pratileira",
    stacked=True, 
    figsize=(10, 5),
    xlabel="Fabricante",
    ylabel="Quantidade",
)
```

```python
df.pivot_table(
    index="mfr",
    columns="type",
    aggfunc="size"
).plot(kind="bar",
    rot=True,
    title="Quantidade de cereais por pratileira",
    figsize=(10, 5),
    xlabel="Fabricante",
    ylabel="Quantidade",
)
```

```python
df.pivot_table(
    index="mfr",
    columns="vitamins",
    aggfunc="size"
).plot(kind="bar",
    rot=True,
    title="Quantidade de cereais por pratileira",
    figsize=(10, 5),
    xlabel="Fabricante",
    ylabel="Quantidade",
)
```
